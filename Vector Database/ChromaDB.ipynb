```python
#Paragraph Tokenization Into Sentences
import nltk
from nltk.tokenize import sent_tokenize

nltk.download('punkt')
nltk.download('punkt_tab') # Added to download the missing resource

text="""The Dance of the Dragons, the catastrophic Targaryen civil war depicted
in House of the Dragon, stands as one of the bloodiest and most tragic
periods in the history of George R. R. Martin’s world. The conflict
began during the reign of King Viserys I Targaryen, who named his
daughter Rhaenyra Targaryen as his heir. Although the lords of Westeros
swore allegiance to her, Viserys later fathered sons with his second
wife, Alicent Hightower, creating a succession crisis. When Viserys
died, Alicent’s supporters—the Greens—crowned Aegon II Targaryen king in
King’s Landing, while Rhaenyra’s supporters—the Blacks—proclaimed her
queen on Dragonstone. Thus began a war that would destroy nearly every
dragon in Westeros and permanently weaken House Targaryen.

The first major bloodshed came when Rhaenyra’s son Lucerys Velaryon
traveled to Storm’s End aboard his young dragon Arrax to secure support
from House Baratheon. There he encountered his uncle Aemond Targaryen,
who rode Vhagar, the largest and oldest living dragon. During a violent
storm over Shipbreaker Bay, Aemond pursued Lucerys. Arrax attacked
first, but Vhagar responded by tearing Arrax apart in midair and killing
Lucerys. This event transformed a political dispute into an irreparable
blood feud.

Seeking revenge, agents known as Blood and Cheese infiltrated the Red
Keep and murdered Prince Jaehaerys, the young son of Aegon II and
Helaena Targaryen. The horrific murder intensified the hatred between
both factions. Across the realm, armies mobilized while dragons became
living weapons of mass destruction.

One of the earliest major dragon battles occurred at Rook’s Rest. The
Blacks sent Rhaenys Targaryen riding Meleys, known as the Red Queen, to
defend the castle. She was ambushed by Aegon II on Sunfyre and Aemond on
Vhagar. Meleys initially overwhelmed Sunfyre, but Vhagar entered the
battle and attacked from above. The combined assault killed Meleys and
Rhaenys. Sunfyre survived but suffered horrific injuries, and Aegon II
was severely burned and crippled.

As losses mounted, Rhaenyra initiated the Sowing of the Seeds, allowing
individuals with Valyrian ancestry to attempt bonding with riderless
dragons. Successful dragonseeds included Addam Velaryon with Seasmoke,
Hugh Hammer with Vermithor, Ulf White with Silverwing, and Nettles with
Sheepstealer.

The Battle of the Gullet became one of the largest naval battles in
Westerosi history. The Triarchy fleet attacked forces loyal to House
Velaryon near Dragonstone. Although the Blacks eventually won, they
suffered devastating losses. Jacaerys Velaryon, Rhaenyra’s eldest son
and heir, was killed while flying Vermax. Vermax crashed and died during
the battle.

Afterward, the Blacks captured King’s Landing, forcing Aegon II to flee.
Rhaenyra finally sat upon the Iron Throne, but her reign quickly
deteriorated because of food shortages, economic collapse, and growing
unrest. Meanwhile, Aemond and Vhagar devastated the Riverlands through a
campaign of destruction.

One of the most legendary dragon duels in history occurred above the
Gods Eye. Daemon Targaryen, riding Caraxes, challenged Aemond and
Vhagar. During the aerial battle, Daemon leaped from Caraxes onto Vhagar
and drove Dark Sister through Aemond’s eye. The dragons crashed into the
lake below. Vhagar died, ending the life of the oldest dragon alive.
Caraxes survived the crash briefly but later died from his injuries.

The most devastating event for dragonkind was the Storming of the
Dragonpit. Driven by fanaticism and hatred of dragons, thousands of
smallfolk attacked the Dragonpit in King’s Landing. Tyraxes, the dragon
of Prince Joffrey Velaryon, was killed while chained. Morghul and
Shrykos, both young dragons, were hacked to death by the attackers.
Dreamfyre, one of the oldest dragons in Westeros, broke free and
slaughtered many rioters. However, she crashed into the structure
supporting the Dragonpit, causing its collapse. Dreamfyre and thousands
of people were buried beneath the falling stone.

The war’s final major battles occurred at Tumbleton. During the First
Battle of Tumbleton, Hugh Hammer and Ulf White betrayed the Blacks,
taking Vermithor and Silverwing to the Green side. During the Second
Battle of Tumbleton, multiple dragons fought simultaneously. Vermithor
killed Seasmoke but was mortally wounded and died. Seasmoke also
perished. Tessarion suffered fatal injuries and died after the battle.

By the end of the war, nearly every major dragon had died. Arrax,
Meleys, Vermax, Vhagar, Caraxes, Tyraxes, Morghul, Shrykos, Dreamfyre,
Seasmoke, Vermithor, and Tessarion were all dead. Sunfyre survived his
earlier wounds but later died after killing Rhaenyra. Only a handful of
dragons remained alive, including Silverwing, Sheepstealer, Cannibal,
and the young Morning.

The war reached its tragic climax when Rhaenyra attempted to flee to
Dragonstone, believing it remained loyal. Instead, she found the island
under Aegon II’s control. Before the eyes of her son Aegon III, Rhaenyra
was fed to Sunfyre. Yet Aegon’s victory was hollow. Most of his allies
were dead, the realm was exhausted, and his support was collapsing. Soon
afterward, Aegon II was poisoned by conspirators seeking peace.

The throne ultimately passed to Aegon III. Although Rhaenyra’s bloodline
continued the Targaryen dynasty, the Dance left Westeros devastated.
Tens of thousands had died, entire regions were ruined, and the
dragons—the source of Targaryen supremacy—were nearly extinct. The war
proved that the greatest threat to House Targaryen was not an external
enemy but the destructive ambition of the family itself.
"""

sentences=sent_tokenize(text)

print(sentences)
```

```python
indexed_sentences=list(enumerate(sentences))
for idx_sentence in indexed_sentences:
  print(idx_sentence)
```

```python
#Now Generating The Embeddings For Each Sentences
from langchain_huggingface import HuggingFaceEndpointEmbeddings
from dotenv import load_dotenv
import os

load_dotenv()
HUGGINGFACEHUB_API_TOKEN = os.getenv("HUGGINGFACEHUB_API_TOKEN")

embeddings = HuggingFaceEndpointEmbeddings(model="BAAI/bge-m3", huggingfacehub_api_token=HUGGINGFACEHUB_API_TOKEN)

sent_embeddings=embeddings.embed_documents(sent[1] for sent in indexed_sentences)
```

## The Most Awaited Topic, Implementing VectorDB Using ChromaDB

```python
# !pip install ChromaDB
```

```python
import chromadb
client=chromadb.Client()
collection=client.create_collection(
    name="HOTD_embeddings"
)
```

```python
#Adding The Data To VectorDB
#Schema--> id | Sentence | Embeddings

collection.add(
    ids=[str(ids[0]) for ids in indexed_sentences],
    documents=[sentence[1] for sentence in indexed_sentences],
    embeddings=sent_embeddings #sentence_embeddings is a list
)
```

```python
#Now Retriving The Documents(Content) From The VectorDB That Best Matches The Users Query
query="Dragons That Died In Targaryen Civil War?"

query_embedding=embeddings.embed_query(query)

#No need to explicitly calculate and compare the similarity scores for embeddings of each document in the DB with the embeddings of query "collection.query()" does it for us
result=collection.query(
    query_embeddings=[query_embedding],
    n_results=20
)

print(result['documents'])
```
