```python
#Cleaning The Sentences
cleaned_sentences=[]
for sentence in sentences:
  new_sentence = sentence.replace("\n", " ")
  cleaned_sentences.append(new_sentence)
```

# Now Generating The Embeddings For The Sentences Using Opensource Embedding Model From HuggingFace "BAAI/bge-3"


```python
!pip install langchain_huggingface
```

```python
from langchain_huggingface import HuggingFaceEndpointEmbeddings
import os
from dotenv import load_dotenv
load_dotenv()

HUGGINGFACEHUB_API_TOKEN = os.getenv("HUGGINGFACEHUB_API_TOKEN")

embedding_model=HuggingFaceEndpointEmbeddings(model="BAAI/bge-m3",huggingfacehub_api_token=HUGGINGFACEHUB_API_TOKEN)

embeddings=embedding_model.embed_documents([sentence[1] for sentence in Indexed_Sentences])
```

```python
#Verifying The Length Embeddings
len(embeddings)
```

```python
#Indexing The Embeddings
Indexed_Embeddings=list(enumerate(embeddings))
```

```python
print(Indexed_Embeddings)
```

# Performing Semantic Chunking On The Embeddings

```python
#getting the cosine similarity of neighbouring sentences
from sklearn.metrics.pairwise import cosine_similarity
i=0
j=1
similarity_scores=[]
while i<len(Indexed_Embeddings)-1 and j<len(Indexed_Embeddings):
    similarity=cosine_similarity([Indexed_Embeddings[i][1]],[Indexed_Embeddings[j][1]])[0][0]
    similarity_scores.append([(i,j,similarity)])
    print(f"Similarity between sentence {i} and sentence {j} is: {similarity}")
    i+=1
    j+=1
```

```python
print(similarity_scores)
```

```python
#Semantic Chunking
import numpy as np
scores=[similarity_score[0][2] for similarity_score in similarity_scores]
mean=np.mean(scores)
std=np.std(scores)
threshold=mean-std
print(threshold)
```

```python
chunks = []
chunk_number = 1
current_chunk = [chunk_number, 0]  # Chunk 1 starts with sentence 0

for score in similarity_scores:
    similarity = score[0][2]
    next_sentence_index = score[0][1]

    if similarity < threshold:
        chunks.append(current_chunk)

        chunk_number += 1
        current_chunk = [chunk_number]

    current_chunk.append(next_sentence_index)

# Add last chunk
chunks.append(current_chunk)

print(chunks)
#format
#[[chunck_number,included_sentences_indexes.....]]
```

```python
# Step 1: Make a simple way to find a sentence using its index
sentence_lookup = {}
for idx_sentence in Indexed_Sentences:
    index = idx_sentence[0]
    sentence = idx_sentence[1]
    sentence_lookup[index] = sentence

# Step 2: Go through each chunk and collect its sentences
sent_chunks = []

for chunk in chunks:
    current_chunk_sentences = []

    for index in chunk[1:]:
        sentence = sentence_lookup[index]
        current_chunk_sentences.append(sentence)

    # join all sentences of this chunk into one paragraph
    full_chunk_text = " ".join(current_chunk_sentences)
    sent_chunks.append(full_chunk_text)
```

#Implementing Vector Database Using ChromaDB

```python
!pip install ChromaDB
```

```python
import chromadb
client=chromadb.Client()

collection=client.create_collection(
    name="HOTD_VECTORDB"
)
```

```python
#Adding The Data To VectorDB
#Schema--> id | Sentence | Embeddings

collection.add(
    ids=[str(i) for i in range(len(sent_chunks))],
    documents=sent_chunks,
    embeddings=embedding_model.embed_documents(sent_chunks)
)
```

```python
#Now Retriving The Documents(Content) From The VectorDB That Best Matches The Users Query
query="Who Killed Vhagar?"

query_embedding=embedding_model.embed_query(query)

#No need to explicitly calculate and compare the similarity scores for embeddings of each document in the DB with the embeddings of query "collection.query()" does it for us
result=collection.query(
    query_embeddings=[query_embedding],
    n_results=3
)

#context
print(result['documents'])
```

# An Enhancement To Standard Rag, Implementing Query Enhancement For Better Semantics To Deal With The Bland User Queries.

```python
from langchain_huggingface import HuggingFaceEndpoint,ChatHuggingFace

llm=HuggingFaceEndpoint(
    repo_id="meta-llama/Llama-3.1-8B-Instruct",
    task="text-generation",
    huggingfacehub_api_token=HUGGINGFACEHUB_API_TOKEN
)

model=ChatHuggingFace(llm=llm)

prompt1=f"""
You are a query rewriting assistant for a Retrieval-Augmented Generation (RAG) system.

Your task is to rewrite the user's query so that it is clear, specific, and optimized for semantic search in a vector database.

Guidelines:
1. Preserve the original intent of the user.
2. Do not answer the question.
3. Do not introduce information not implied by the query.
4. Return only the rewritten query and nothing else.
5.Do Not Add Any Extra Information To The Query From Your End Just Improve The Semantics Of The Query
6.You Know Nothing About The Context Of The Query

User Query:
{query}

Rewritten Query:
"""
result1=model.invoke(prompt1)

enhanced_query=result1.content
```
