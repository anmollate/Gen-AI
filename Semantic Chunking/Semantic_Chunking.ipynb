```python
import nltk
from nltk.tokenize import sent_tokenize

nltk.download('punkt')

with open("The_Last_Lantern_of_Willow_Creek.txt", "r", encoding="utf-8") as file:
    text = file.read()

sentences = sent_tokenize(text)

print(sentences)
```

```python
sentences=[sentence.replace("\n","") for sentence in sentences]
```

```python
indexed_sentences=list(enumerate(sentences))
print(indexed_sentences)
```

```python
print(indexed_sentences[0][1])
```

```python
#generating emebeddings for the sentences
from langchain_huggingface import HuggingFaceEndpointEmbeddings
from dotenv import load_dotenv

load_dotenv()

embeddings = HuggingFaceEndpointEmbeddings(model="BAAI/bge-m3")

chunks_embedding = embeddings.embed_documents([s[1] for s in indexed_sentences])

print(len(chunks_embedding))



```

```python
chunks_embedding=list(enumerate(chunks_embedding))
print(chunks_embedding[0])
```

```python
#CSV file Creation For Displaying The Sentences Corresponding To Their Embeddings
import pandas as pd
data=[]
for sentence in indexed_sentences:
    sentence_index=sentence[0]
    embedding=chunks_embedding[sentence_index][1]
    data.append({
        "sentence":sentence[1],
        "embedding":embedding
    })
df = pd.DataFrame(data)
df.to_csv("sentences_with_embeddings.csv", index=False)
```

```python
#getting the cosine similarity of neighbouring sentences
from sklearn.metrics.pairwise import cosine_similarity
i=0
j=1
similarity_scores=[]
while i<len(chunks_embedding)-1 and j<len(chunks_embedding):
    similarity=cosine_similarity([chunks_embedding[i][1]],[chunks_embedding[j][1]])[0][0]
    similarity_scores.append([(i,j,similarity)])
    print(f"Similarity between sentence {i} and sentence {j} is: {similarity}")
    i+=1
    j+=1

```

```python
print(similarity_scores)
print(len(similarity_scores))
```

```python
#Calculating The Threshold for similarity scores where threshold = mean - std
import numpy as np
scores=[similarity_score[0][2] for similarity_score in similarity_scores]
mean=np.mean(scores)
std=np.std(scores)
threshold=mean-std
print(threshold)
```

```python
#Now grouping the chunks of sentences based on their similarity scores and the threshold
#for ex: if similarity between 1 and 2>0.42 group them, if similarity between 2 and 3>0.42 group them with 1 and 2 and so on, if the threshold value is not met then start a new group of chuncks
chunks = []
chunk_number = 1
current_chunk = [chunk_number, 0]  # Chunk 1 starts with sentence 0

for score in similarity_scores:
    similarity = score[0][2]
    next_sentence_index = score[0][1]

    if similarity < threshold:
        chunks.append(current_chunk)

        chunk_number += 1
        current_chunk = [chunk_number]

    current_chunk.append(next_sentence_index)

# Add last chunk
chunks.append(current_chunk)

print(chunks)
```

```python
#Summarizing the above grouping of sentences into chuncks with respect to their semantic meaning, using csv file
import pandas as pd

data=[]
for chunk in chunks:
    chunk_number=chunk[0]
    
    for sentence_index in chunk[1:]:
        data.append({
            "chunk_number":chunk_number,
            "sentence":indexed_sentences[sentence_index][1]
        })

df = pd.DataFrame(data)
df.to_csv("grouped_chunks.csv", index=False)
```
